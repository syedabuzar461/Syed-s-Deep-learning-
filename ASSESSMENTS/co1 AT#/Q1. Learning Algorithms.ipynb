```
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

m, c = np.polyfit(x, y, 1)
pred = m*x + c

loss = np.mean((y-pred)**2)
print("Slope:", round(m,2))
print("Intercept:", round(c,2))
print("Loss:", round(loss,4))

plt.plot(x,y,'o')
plt.plot(x,pred)
plt.show()
```
